# Matplotlib Plotting Time Series Data

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
```

```python
plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates, y)
```

```python
plt.plot_date(dates, y, linestyle="solid")

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```

```python
plt.plot_date(dates, y, linestyle="solid")
plt.gcf().autofmt_xdate()               # gcf- get current figure,  autofmt - auto format

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```

```python
plt.plot_date(dates, y, linestyle="solid")
plt.gcf().autofmt_xdate()               # gcf- get current figure,  autofmt - auto format

date_format = mpl_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)
# get current axis

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```

# Working on CSV TimeDate data


```python
data = pd.read_csv('data/time_series.csv')
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)
```

```python
data.head()
```

```python
price_date = data['Date']
price_close = data['Close']
```

```python
plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```

```python
plt.plot_date(price_date, price_close, linestyle='solid')
plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)
# get current axis

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```
