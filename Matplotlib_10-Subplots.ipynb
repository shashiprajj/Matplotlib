# Matplotlib Subplots

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.style.use('seaborn')

data = pd.read_csv('data/subplot.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, py_salaries, label='Python')
plt.plot(ages, js_salaries, label='JavaScript')

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
```

```python
plt.plot(ages, py_salaries, label='Python')
plt.plot(ages, js_salaries, label='JavaScript')

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
plt.style.use('seaborn')

data = pd.read_csv('data/subplot.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# Creating objects of Subplot, so that for future we can have multiple objets for multiple graphs
fig, ax = plt.subplots()

ax.plot(ages, py_salaries, label = "Python")
ax.plot(ages, js_salaries, label = "JavaScript")

ax.plot(ages, dev_salaries, color="#444444", linestyle="--", label ="All Devs")

ax.legend()
ax.set_title("Median Salary (USD) by Age")
ax.set_xlabel("Ages")
ax.set_ylabel("Median Salary (USD)")
plt.tight_layout()

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)


ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 25)
y = x*x+2
```

```python
fig, Axes = plt.subplots(1, 2, figsize=(10,8))          # 1 row, 2 columns

Axes[0].plot(x, x**2, x**3, lw=2)
Axes[0].grid(True)

Axes[1].plot(x, x**2, x**3, lw=2)
Axes[1].set_ylim([0, 60])
Axes[1].set_xlim([2, 5])          # to see where it crosses both the lines of graph by zooming or setting limit on x and y axis
Axes[1].grid(True)
```

# Other 2D Graphics

```python
n = np.array([0, 1,2,3,4,5])
```

```python
fig, Axes = plt.subplots(1, 4, figsize=(15,5))

Axes[0].set_title("Scatter")
Axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))
```

```python
fig, Axes = plt.subplots(1, 4, figsize=(15,5))

Axes[0].set_title("Scatter")
Axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))

Axes[1].set_title("Step")
Axes[1].step(n, n*2, lw=2)
```

```python
fig, Axes = plt.subplots(1, 4, figsize=(15,5))

Axes[0].set_title("Scatter")
Axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))

Axes[1].set_title("Step")
Axes[1].step(n, n*2, lw=2)

Axes[2].set_title("Bar")
Axes[2].bar(n, n*2, align='center', lw=2)
```

```python
fig, Axes = plt.subplots(1, 4, figsize=(15,5))

Axes[0].set_title("Scatter")
Axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))

Axes[1].set_title("Step")
Axes[1].step(n, n*2, lw=2)

Axes[2].set_title("Bar")
Axes[2].bar(n, n*2, align='center', lw=2)

Axes[3].set_title("fill Between")
Axes[3].fill_between(x, x**2, x**3, color='green')
```
